```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/StudentsPerformance.csv')
```

```python
new_row={'gender':'female','race/ethnicity':'group A','parental level of education':'some college','lunch':'standard','test preparation course':'none','math score':65,'writing score':91}
df.loc[1000]=new_row
df
```

```python
df.fillna(np.mean(df,axis=0),inplace=True)
```

```python
df
```

```python
xpoints = np.array([0,34,70])
ypoints = np.array([0,24,50])

plt.figure(figsize=(15,7))
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.plot(xpoints,ypoints,marker='o')
plt.title('coordinates')
plt.show()
```

```python
xpoints = np.array([0,34,70])
ypoints = np.array([0,24,50])

xpoints1 = np.array([0,24,56])
ypoints1 = np.array([0,72,21])

plt.scatter(xpoints,ypoints)
plt.scatter(xpoints1,ypoints1)
plt.show()
```
